```python
import pandas as pd
data=pd.read_csv("C:/Users/25160/Desktop/ginf.csv")
d1=data[(data["league"]=="D1")&(data["season"]==2012)]
d1 = d1.reset_index(drop=True)
l1=list(d1["ht"])+list(d1["at"])
l1=list(set(l1))
a1=dict(zip(l1,[0 for i in l1]))#赢
b1=dict(zip(l1,[0 for i in l1]))#平
c1=dict(zip(l1,[0 for i in l1]))#败
r1=[]
```

```python
for i in range(d1.shape[0]):
    if d1.loc[i]['fthg']>d1.loc[i]['ftag']:
        a1[d1.loc[i]['ht']]+=1
        c1[d1.loc[i]['at']]+=1
    elif d1.loc[i]['fthg']<d1.loc[i]['ftag']:
        a1[d1.loc[i]['at']]+=1
        c1[d1.loc[i]['ht']]+=1
    else:
        b1[d1.loc[i]['ht']]+=1
        b1[d1.loc[i]['at']]+=1
for i in range(len(a1)):
    r1.append([list(a1.items())[i][0],list(a1.items())[i][1],list(b1.items())[i][1],list(c1.items())[i][1]])
r1=pd.DataFrame(r1)
r1.columns=['name','a','b','c']
#r1.to_csv('C:/Users/25160/Desktop/2.csv', index=False)
```
