```python
import pandas as pd
data=pd.read_csv("C:/Users/25160/Desktop/ginf.csv")
d1=data[(data["league"]=="D1")&(data["season"]==2012)]
d1 = d1.reset_index(drop=True)
l1=list(d1["ht"])+list(d1["at"])
l1=list(set(l1))
t1=dict(zip(l1,[0 for i in l1]))#进球数
s1=dict(zip(l1,[0 for i in l1]))#丢球数
r1=[]
r2=[]
for i in range(d1.shape[0]):
    t1[d1.loc[i]['ht']]+=int(d1.loc[i]['fthg'])
    t1[d1.loc[i]['at']]+=int(d1.loc[i]['ftag'])
    s1[d1.loc[i]['ht']]+=int(d1.loc[i]['ftag'])
    s1[d1.loc[i]['at']]+=int(d1.loc[i]['fthg'])
for i in range(len(t1)):
    r1.append([list(t1.items())[i][0],list(t1.items())[i][1]])
    r2.append([list(t1.items())[i][0],list(s1.items())[i][1]])
r1
```

```python
r1=pd.DataFrame(r1)
r2=pd.DataFrame(r2)
```

```python
r1.columns=['name','goal']
r2.columns=['name','lose']
r1.to_csv('C:/Users/25160/Desktop/3_1.csv', index=False)
r2.to_csv('C:/Users/25160/Desktop/3_2.csv', index=False)
```
